## Fetch data from database and add them into testset

```python
import pymysql
list2 =[]
conn=pymysql.connect(host='127.0.0.1', user='root', password='', port=3306, database='projectdb')
dbcur = conn.cursor()
dbcur.execute("select * from projectdata")
for db in dbcur:
    list2.append(list(db))
    print(db)
```

## For numerical variable, fill null values with their respective mean or median
## For categorical variable, fill null values with their respective mode

```python
test2['Gender'].fillna(test2['Gender'].mode()[0], inplace=True)
test2['Dependents'].fillna(test2['Dependents'].mode()[0], inplace=True)
test2['Self_Employed'].fillna(test2['Self_Employed'].mode()[0], inplace=True)
test2['LoanAmount'].fillna(test2['LoanAmount'].median(), inplace=True)
test2['Loan_Amount_Term'].fillna(test2['Loan_Amount_Term'].mode()[0], inplace=True)
test2['Credit_History'].fillna(test2['Credit_History'].mode()[0], inplace=True)
test2['Property_Area'].fillna(test2['Property_Area'].mode()[0], inplace=True)
```

```python
test2.isnull().sum()
```

```python
test2['LoanAmount_log'] = np.log(test2['LoanAmount'])
```

```python
test3 = test2.drop('Loan_ID', axis=1)
```

```python
test3 = pd.get_dummies(test3)
test3.head()
```

```python
import pickle
model = pickle.load(open("finalized_model.sav", 'rb'))
```

```python
pred_test = model.predict(test3)
```

```python
submission = pd.DataFrame()
```

```python
submission['Loan_ID'] = test2['Loan_ID']
submission['Loan_Status'] = pred_test
```

```python
submission['Loan_Status'].replace(0, 'N', inplace=True)
submission['Loan_Status'].replace(1, 'Y', inplace=True)
```

```python
pd.DataFrame(submission, columns=['Loan_ID', 'Loan_Status']).to_csv('output_logistic.csv')
```

## Add these predicted values in database table

```python
import pymysql
conn2 = pymysql.connect(host='127.0.0.1', user='root', password='', port=3306, database='projectdb')
mycur = conn2.cursor()
```

```python
try:
    mycur.execute("create table sub_table2(loan_id varchar(20) primary key, loan_status varchar(10))")
except:
    None
```

```python
mycur.execute("show tables")
for x in mycur:
    print(x)
```

```python
with open('output_logistic.csv', 'r') as op:
    for lines in op:
        attrArray = lines.split(",")
        list1 = (attrArray[1], attrArray[2])
        if 'Y' in list1[1]:
            try:
                mycur.execute("insert into sub_table2(loan_id, loan_status) values(%s, %s)", list1)
            except:
                continue
        conn2.commit()         
```

```python
mycur.execute("select * from sub_table2")
for x in mycur:
    mycur.execute("update sub_table2 SET loan_status = TRIM(TRAILING '\n' FROM loan_status);")
```
